```python
# Import dataset library
import tensorflow_datasets as tfds

print("Loading Oxford Flowers 102 dataset...")
print("(First time will download ~300MB - takes 1-2 minutes)")
```

```python
# Load the dataset (this will download on first run)
dataset, dataset_info = tfds.load(
    'oxford_flowers102',
    with_info=True,
    as_supervised=True
)

print("✅ Dataset loaded successfully!")
```

```python
# Display dataset information
print(dataset_info)
```

```python
# Split the dataset into train, validation, and test
train_set = dataset['train']
val_set = dataset['validation']
test_set = dataset['test']

print(f"✅ Training samples: 1,020")
print(f"✅ Validation samples: 1,020")
print(f"✅ Test samples: 6,149")
print(f"✅ Number of classes: 102")
```

```python
# Import visualization library
import matplotlib.pyplot as plt

print("✅ Matplotlib imported - ready to visualize!")
```

```python
# Take one image from the training set
for image, label in train_set.take(1):
    plt.figure(figsize=(6, 6))
    plt.imshow(image.numpy())
    plt.title(f"Flower Label: {label.numpy()}")
    plt.axis('off')
    plt.show()
    
    print(f"Image shape: {image.shape}")
    print(f"Label: {label.numpy()}")
```

```python
# Display 9 flower images in a grid
plt.figure(figsize=(12, 12))

for i, (image, label) in enumerate(train_set.take(9)):
    plt.subplot(3, 3, i + 1)
    plt.imshow(image.numpy())
    plt.title(f"Label: {label.numpy()}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
